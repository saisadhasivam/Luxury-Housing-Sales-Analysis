```python
import pandas as pd
import os

BASE_DIR = os.getcwd()
CSV_PATH = os.path.join(
    BASE_DIR,
    "..",
    "01_raw_data",
    "Luxury_Housing_Bangalore.csv"
)

df = pd.read_csv(CSV_PATH)

df.head()
```

```python
# Column-wise non-null count
df.info()
```

```python
# Missing values per column
missing_counts = df.isnull().sum()
missing_counts
```

```python
# Missing value percentage per column
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.round(2)
```

```python
# Columns with missing values only
missing_summary = pd.concat(
    [missing_counts, missing_percent],
    axis=1,
    keys=["Missing_Count", "Missing_Percentage"]
)

missing_summary[missing_summary["Missing_Count"] > 0].sort_values(
    by="Missing_Percentage",
    ascending=False
)
```

### EDA Conclusions

Key data quality observations from raw dataset:

1. Missing Values:
   - Buyer_Comments: ~18%
   - Amenity_Score: ~10%
   - Unit_Size_Sqft: ~10%
   - Ticket_Price_Cr: ~10%

2. Data Type Issues:
   - Ticket_Price_Cr stored as text with currency symbols and mixed formats
   - Purchase_Quarter stored as string instead of datetime

3. Text Standardization Issues:
   - Micro_Market values contain casing and spelling variations
   - Configuration values are inconsistent (3Bhk, 3bhk, 3BHK, etc.)

4. Duplicate Records:
   - ~1000 duplicate Property_IDs detected

5. Invalid Numeric Values:
   - Unit_Size_Sqft contains negative values (-1)
   - Some unit sizes exceed realistic thresholds
